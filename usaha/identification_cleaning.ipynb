### Transformasi Kolom KBLI

### Menghitung Jumlah Data per golongan_kbli

```python
def count_column_values(df, column):
    uniqueValues = cluster_df[column].unique()
    objectCount = {
        'KodeGolonganKBLI': [],
        'jumlahData': [],
    }
    for uniqueValue in uniqueValues:
        uniqueValueCount = len(df[column][df[column] == uniqueValue])
        objectCount['KodeGolonganKBLI'].append(uniqueValue)
        objectCount['jumlahData'].append(uniqueValueCount)
    result_df = pd.DataFrame(objectCount)
    result_df.set_index('KodeGolonganKBLI', inplace=True)
    
    result_df.sort_values(by=['jumlahData'], ascending=False, inplace=True)
    return result_df
```

#### Daftar golongan_kbli yang terdapat pada dataset:

```python
golongan_kbli_count_unique = len(cluster_df.golongan_kbli.unique())
print("Jumlah golongan KBLI yang terdapat pada dataset: ", golongan_kbli_count_unique, 'Kode Golongan KBLI')
```

```python
golongan_unique_df = count_column_values(cluster_df, 'golongan_kbli')
golongan_unique_df
```

### Filter Kode Golongan KBLI

### Filter dataset dengan Kode Golongan KBLI yang melebihi 50 data

```python
def filter_dataframe_by_values(dataframe, column_name, values_to_include):
    if column_name not in dataframe.columns:
        print(f"Kolom / Atribut '{column_name}' Tidak ditemukan dalam dataframe.")
        return None
    
    filtered_dataframe = dataframe[dataframe[column_name].isin(values_to_include)].copy()
    return filtered_dataframe
```

```python
filtered_cluster_df = filter_dataframe_by_values(cluster_df, 'golongan_kbli', filtered_golongan_unique_df.index.values)
```

# Transformasi Data untuk Clustering

Melakukan transformasi terhadap dataset cluster sehingga tipe dari dataset menjadi numerik.<br> Dataset baru hasil transformasi dibuat dengan menghitung jumlah label pada setiap atribut/kolom dan melakukan pengelompokkan per kecamatan.

```python
def count_coulumn_values(df, location, column, col_values):
    count_values = len(df[(df['kecamatan_usaha'] == location) & (df[column] == col_values)])
    return count_values
```

```python
def transformClusteringDataset(df, column):
    df_target_column = df[[column, 'kecamatan_usaha']]
    
    # buat dataframe kosong, kelompokkan berdasarkan kecamatan_usaha
    df_transform = pd.DataFrame({
        'kecamatan': df_target_column.kecamatan_usaha.unique()
    })
    df_transform = df_transform.set_index('kecamatan')
    
    # buat kolom/atribut berdasarkan label dari kolom target 
    for unique_value in df_target_column[column].unique():
        df_transform[unique_value] = 0
        
    # isi nilai setiap kolom
    for index, row in df_transform.iterrows():
        for col in df_transform.columns:
            df_transform.loc[index , col] = count_coulumn_values(df_target_column, index, column, col)
    
    return df_transform
```

Kolom/Atribut Clustering

```python
final_target_cluster_columns = [
    'uraian_jenis_perusahaan',
    'uraian_risiko_proyek',
    'skala_usaha',
    'golongan_kbli',
]
```

#### Simpan Hasil Transformasi Dataset

```python
for cluster_column in final_target_cluster_columns:
    df_transform = transformClusteringDataset(filtered_cluster_df, cluster_column)
    filename = 'transform_df_' + cluster_column + '.csv'
    df_transform.to_csv('transform_datasets/' + filename)
```
